```python
import pandas as pd
```

```python
df = pd.read_csv('gov_data_all.csv', index_col = False)
```

```python
arr = sorted(df)
for i in range(len(arr)):
    arr[i] = arr[i].split('.')
    if 'assoc' in arr[i]:
        arr[i] = arr[i][arr[i].index('assoc'):]
    elif 'program_percentage' in arr[i]:
        arr[i] = arr[i][arr[i].index('program_percentage'):]
    elif 'race_ethnicity' in arr[i]:
        arr[i] = arr[i][arr[i].index('race_ethnicity'):] 
    elif 'act_scores' in arr[i]:
        arr[i] = ['ACT Average']
    elif 'sat_scores' in arr[i]:
        arr[i] = ['SAT Average']
    elif 'admission_rate' in arr[i]:
        arr[i] = ['admission rate']
    elif 'cost' in arr[i]:
        arr[i] = arr[i][arr[i].index('cost'):]
    elif 'earnings' in arr[i]:
        arr[i] = arr[i][arr[i].index('earnings'):]
    elif 'demographics' in arr[i]:
        arr[i] = arr[i][arr[i].index('demographics'):]
    elif 'retention_rate' in arr[i]:
        arr[i] = arr[i][arr[i].index('retention_rate'):]
    elif 'retention_rate_suppressed' in arr[i]:
        arr[i] = arr[i][arr[i].index('retention_rate_suppressed'):]
    elif 'location' in arr[i]:
        arr[i] = arr[i][arr[i].index('location')+1:]
    elif 'school' in arr[i]:
        arr[i] = arr[i][arr[i].index('school')+1:]
        if 'name' in arr[i]:
            arr[i] = 'REMOVE'
    elif 'cip_4_digit' in arr[i]:
        arr[i] = ['programs cip_4_digit']
    elif 'student' in arr[i]:
        arr[i] = ['student size']
    elif 'True' in arr[i]:
        arr[i] = ['name']
        
    arr[i] = ' '.join(arr[i])
    
arr
```

```python
# prints the names when they don't line up (should be fine because we changed the names a lot for a few)
cols = sorted(df)
for i in range(len(cols)):
    if not '.'.join(arr[i].split()) in cols[i]:
        print(arr[i])
        print(cols[i])
        print()
```

```python
d = {}
for i in range(len(cols)):
    d[cols[i]] = arr[i]

    
df = df.rename(columns=d)
df = df.drop('R E M O V E', axis=1)
df
```
